```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), color='orange')
plt.plot(x, np.sin(x-1), color='b')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), linestyle='solid')
plt.plot(x, np.sin(x-1), linestyle='dashed')
plt.plot(x, np.sin(x-2), linestyle='dashdot')
plt.plot(x, np.sin(x-3), linestyle='dotted')

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), linestyle='-')
plt.plot(x, np.sin(x-1), linestyle='--')
plt.plot(x, np.sin(x-2), linestyle='-.')
plt.plot(x, np.sin(x-3), linestyle=':')

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), '-')
plt.plot(x, np.sin(x-1), '--')
plt.plot(x, np.sin(x-2), '-.')
plt.plot(x, np.sin(x-3), ':')

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), 'orange')
plt.plot(x, np.sin(x-1), 'b')
plt.plot(x, np.sin(x-2), '0.75')
plt.plot(x, np.sin(x-3), '#FFDD44')

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x-0), 'r-')
plt.plot(x, np.sin(x-1), '--g')
plt.plot(x, np.sin(x-2), '-.y')
plt.plot(x, np.sin(x-3), ':b')

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x))
plt.xlim(-0.5, 5.5)
plt.ylim(-1.2, 1.2)

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x))
plt.xlim(5.5, -0.5)
plt.ylim(-1.2, 1.2)

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x))
plt.axis([-0.5, 5.5, -1.2, 1.2])
# plt.xlim(5.5, -0.5)
# plt.ylim(-1.2, 1.2)

plt.show()
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, np.sin(x), '-p',
         markersize=20,
         markerfacecolor='white',
         markeredgewidth=3
)
plt.axis([-0.5, 5.5, -1.2, 1.2])

plt.show()
```

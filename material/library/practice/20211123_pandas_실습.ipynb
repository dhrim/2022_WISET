````python
df = pd.DataFrame(
    {"AAA":[4,5,6,7],
     "BBB":["남", "여", "", "여"],
     "CCC":[100,50,-30,-50]}
)

def encode(BBB):
  if BBB=="남": return 1
  if BBB=="여": return 0
  return -1

df['BBB_encoded'] = df[['BBB']].applymap(encode)
print(df)

new_df = df.drop(columns="BBB")
print(new_df)
print(df)

df.drop(columns="BBB", inplace=True)
print(df)
````

```
numpy로 0~1 랜덤한 값을 10개 만들고

10개의 값을 가지는 r 컬럼의 dataframe을 만들고

five_r이라는 r 컬럼에 5배 곱한 컬럼을 새로 만들고

이를 정수화 한 값의 cat 이라는 이름의 컬럼을 만든다.

cat외의 모든 컬럼을 삭제한다.
```





````python
import numpy as np
import pandas as pd

a = np.random.rand(10)
# print(a)

df = pd.DataFrame({"r":a})
# print(df)

df['five_r'] = df.r*5
# print(df)

df['cat'] = df.five_r.astype(np.int16)
# print(df)

df.drop(columns=["r", "five_r"], inplace=True)
print(df)
````

````python
df = pd.DataFrame(
    {"AAA":[4,5,6,7],
     "BBB":[10,20,30,40],
     "CCC":[100,50,-30,-50]}
)

df['float_AAA'] = df.AAA.astype(np.float)
print(df)

df.drop(columns="AAA", inplace=True)
print(df)

df.rename(columns={"float_AAA":"AAA"}, inplace=True)
print(df)
````
